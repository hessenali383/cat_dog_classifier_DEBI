```python
import gradio as gr
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.applications.resnet50 import preprocess_input
from PIL import Image
import tensorflow as tf
```

```python
file_id = '1WRv9OuTsmvOdcGH0Jz1uMHJrdNqOaKfZ'
download_url = f'https://drive.google.com/uc?export=download&id={file_id}'

model_path = tf.keras.utils.get_file(
    'cat_dog_model.h5',
    download_url,
    cache_dir='./',
    cache_subdir='models' )
```

```python
model = load_model(model_path)

class_names = ['Cat', 'Dog']

def predict_image(image):
    image = image.resize((224, 224))
    image = np.array(image)
    image = preprocess_input(image)

    # Add batch dimension
    image = np.expand_dims(image, axis=0)

    prediction = model.predict(image)[0][0]

    if prediction < 0.5:
        return class_names[0]
    else:
        return class_names[1]


  # Create a Gradio interface
app = gr.Interface(fn=predict_image, inputs=gr.Image(type="pil"), outputs=gr.Label(), title="Cat vs Dog Classifier" , description="Upload an image of a cat or dog to see the prediction.")



app.launch()
```
